```python
import os
from PIL import Image,ImageFont,ImageDraw
import xlrd
import pandas as pd


class poster:
    def __init__(self):
        _fn='2019科学实验课学员档案.xlsx'
        self.pth=os.getcwd()
        self.fn=os.path.join(self.pth,_fn)
    
    def rd_excel(self):
        title='吸尘器'
        df=pd.read_excel(self.fn)        
        cont=df.loc[df['实验主题']==title]['知识点']
        out=[title,cont[0]]

        return out
        
    def toimage(self):
        pth=self.pth+'/pics'        
        txt=self.rd_excel()

        #背景图
        bgFile=os.path.join(pth,'bg_white.jpg')
        pic_bg=Image.open(bgFile)
        w_bg,h_bg=pic_bg.size

        
        #课程图       
        crs=os.path.join(pth,'jingdian.jpg')
        pic_crs=Image.open(crs)
        w_crs,h_crs=pic_crs.size
        r_crs=w_crs/h_crs
        new_w=int(w_bg-20)
        new_h=int(new_w/r_crs)        
        pic_new=pic_crs.resize((new_w,new_h))        
        pic_bg.paste(pic_new,(10,10))
        
        
        #课程描述
        fontpath = "msyh.ttc"
        font = ImageFont.truetype(fontpath, 32)
        draw = ImageDraw.Draw(pic_bg) 
        
        txtX=self.posX(txt[0],32,w_bg,'center')        
        draw.text((txtX,new_h+40), txt[0],font=font, fill = (0, 255 ,0))
        
        txtX=self.posX(txt[1],32,w_bg,'center')  
        draw.text((txtX,new_h+80), txt[1],font=font, fill = (255, 0 ,0))
        
        #二维码
        qrFile=os.path.join(pth,'大智小超二维码.jpg')
        pic_qr=Image.open(qrFile)
        w_qr,h_qr=pic_qr.size
        w_qr_new=90
        h_qr_new=90
        pic_qr_edit=pic_qr.resize((w_qr_new,h_qr_new))
        pic_bg.paste(pic_qr_edit,(600-w_qr_new-10,800-h_qr_new-10))
        
        fontpath = "simsun.ttc"
        font = ImageFont.truetype(fontpath, 18)
        draw = ImageDraw.Draw(pic_bg) 
        draw.text((200,740), "长按识别二维码",font=font, fill = '#075B90')
        draw.text((200,765), "关注【大智小超】公众号",font=font, fill = '#075B90')
#         pic_qr_edit.show()
        
        savefn=os.path.join(pth,'out.png')
#         pic_bg.save(savefn)
        pic_bg.show()
        
    def posX(self,t,font_size,w,p):
        #如有回车，计算最长的一行        
        txts=t.split("\n")
        length=[]
        for txt in txts:
            length.append(len(txt))
        max_len=max(length)                  

        
        #输出坐标X值
        if p=='left':
            return 0
        elif p=='right':
            return int(w-font_size*max_len)
        elif p=='center':    
            _w=int((w-font_size*max_len)/2)
            return _w

        

if __name__=='__main__':
    myposter=poster()
    myposter.rd_excel()
    myposter.toimage()
```
