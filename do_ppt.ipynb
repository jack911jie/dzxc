```python
import os
import copy
from pptx import Presentation
from pptx.util import Pt,Inches
from PIL import Image

class split_pptx:
    def __init__(self,*num):
        self.num=[u-1 for u in num] #处理输入的数据，以0为开始
        self.pcs=int(len(self.num))+1
        
    def split(self,ppt):
        self.fn=ppt
        self.prs=Presentation(ppt)
        slides=self.prs.slides
        page_max=len(slides)-1
        
        print("PPT的最大页码为：{}".format(page_max+1)) #显示要以1为开始
        if int(self.num[-1])<=len(slides): #输入的最大页码应小于等于ppt的最大页码  
            if self.num[-1]!=page_max:
                self.num.append(page_max)
                
            print(self.num)
            
            list_remain=[]
            n=0
            for i in self.num:
                _list=[]
                for j in range(n,int(i)):
                    _list.append(j)
                    n+=1
                list_remain.append(_list)
                
            list_remain[-1].append(self.num[-1])#包括最后一页
            
            list_del=self.to_del(list_remain)
            self.del_save(list_remain,list_del)
            
        else:
            print("输入的最大页码不应大于PPT的最大页码")
            input("")
            exit()
            
            
    def del_save(self,p_to_remain,p_to_del):
        n=0
        for arrs in p_to_del:               
            prs_to_del=Presentation(self.fn)
            fn=self.fn[0:-5]+"["+str(p_to_remain[n][0]+1)+"-"+str(p_to_remain[n][-1]+1)+"]"+".pptx"
            slist = prs_to_del.slides._sldIdLst
            # 字典生成式
            pages = {i : sld for i, sld in enumerate(prs_to_del.slides._sldIdLst)}            
            for arr in arrs:                
                slist.remove(pages[arr])
#                 print("第{0}个文件，文件名:{2} ：删除页码为 {1} \n".format(n+1,arr+1,fn))
            print("已完成第{0}个文件，文件名:{1} \n".format(n+1,fn))
            n+=1
            prs_to_del.save(fn)
                             
    def mkdir(self):
        pass
    
    def to_del(self,arr):
        list=[]
#         print(arr)
        for i in range(0,len(arr)):
            _arr=copy.deepcopy(arr) #深拷贝,否则会出错。         
            _arr.pop(i)            
#             print(i,_arr)
            _list=[]
            for j in _arr:
                for k in j:
                    _list.append(k)
                    
            list.append(_list)
#         print(list)
        return list
        
class img_to_ppt:
    def __init__(self,p):
        self.p=p
        
    def imgIn(self):
        prs = Presentation()
        
        for pic in os.listdir(self.p):
            if pic[-3:].lower()=='jpg':

                img_path = os.path.join(p,pic)  # 文件路径
                # 改图片大小
#                 print(os.path.join(self.p,pic))
#                 _img=Image.open(img_path)
#                 ht,wt=_img.size
#                 img=_img.resize((1024,int(1024*wt/ht)))
#                 img.save(img_path)               
                
                left, top, width, height = Inches(0), Inches(1), Inches(10), Inches(2)  # 预设位置及大小  
                slide = prs.slides.add_slide(prs.slide_layouts[6]) #prs.slide_layouts中一共预存有1-48种，采用第6六种为空白幻灯片
                pic = slide.shapes.add_picture(img_path, left, top,width)  # 在指定位置按预设值添加图
        prs.save(os.path.join(self.p,'pic.pptx'))
        print("done")

if __name__=="__main__":
    a="//home//jack//文档//大智小超//乐高//test.pptx"
    p="//home//luo//data//大智小超//乐高//课程//蝌蚪青蛙"
    spt=img_to_ppt(p)
    spt.imgIn()
```
