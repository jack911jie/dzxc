```python
import os
import logging
import pandas as pd
from PIL import Image,ImageDraw,ImageFont

logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(funcName)s-%(lineno)d - %(message)s')
logger = logging.getLogger(__name__)

class poster:
    def __init__(self):
        self.bg='I:\\乐高\\学员上课总结'
        self.font_dir='C:\Windows\Fonts'
        self.default_font='msyh.ttc'
        self.picWid=425
        self.crsList='I:\\乐高\\图纸\\Wedo课程表.xlsx'
    
    def PosterDraw(self,crs_name):
        def basic_para():
            picWid=self.picWid
            r=0.75
            picX1,picX3=10,10
            picX2,picX4=picX1+picWid+20,picX1+picWid+20
            picY1,picY2=770,770
            picY3,picY4=picY1+picWid*r+20,picY1+picWid*r+20

            pic0=pic_xy(picWid,int(picWid*r),440,290) #标题图
            pic1=pic_xy(picWid,int(picWid*r),picX1,picY1) #照片一
            pic2=pic_xy(picWid,int(picWid*r),picX2,picY2) #照片二
            pic3=pic_xy(picWid,int(picWid*r),picX3,picY3) #照片三
            pic4=pic_xy(picWid,int(picWid*r),picX4,picY4) #照片四
            
            return[picWid,picX1,picX2,picX3,picX4,picY1,picY2,picY3,picY4,pic0,pic1,pic2,pic3,pic4]   
        def read_excel():
            df=pd.read_excel(self.crsList)        
            crs=df.loc[df['课程名称']==crs_name]   
            knowledge=list(crs['知识点'])
            script=list(crs['话术'])
            dif_level=list(crs['难度'])
            crs_info=[crs_name,knowledge[0],script[0],dif_level[0]]            
            stars=crs_info[-1].replace('*','★')
            crs_info[-1]=stars 
            
            logging.info('\n'.join(crs_info))                    
            return(crs_info)        
        def expScript():
            name='李XX'
            script=name+'在 【'+crs_info[0]+'】 这节课中，'+crs_info[2]+'\n非常棒！'
            
            return script           
        
        def pic_xy(picWid,picHig,x0,y0):
            #白色矩形
            recX0=x0
            recY0=y0
            recX1=x0+picWid+10
            recY1=y0+picHig+10

            #照片
            picX0=recX0+5
            picY0=recY0+5
            picX1=picX0+picWid
            picY1=picY0+picHig

            return [[int(recX0),int(recY0),int(recX1),int(recY1)],[int(picX0),int(picY0),int(picX1),int(picY1)]]        
        def basic_bg():
            s1=150
            s2=100
            s3=500
            s4=700
            s5=350
            s6=150
            sprt=5      
            total_len=s1+s2+s3+s4+s5+s6+sprt*5

            y1=s1

            y2=y1+sprt        
            y2_2=y2+s2

            y3=y2_2+sprt
            y3_2=y3+s3

            y4=y3_2+sprt
            y4_2=y4+s4

            y5=y4_2+sprt
            y5_2=y5+s5

            y6=y5_2+sprt
            y6_2=y6+s6       

            img = Image.new("RGB",(900,total_len),(255,255,255))
            draw=ImageDraw.Draw(img)
            draw.rectangle([(0,0),(900,y1)],fill='rgb(255,255,255)') #乐高机器人课
            draw.rectangle([(0,y2),(900,y2_2)],fill='rgb(250,254,255)') # 姓名 年龄
            draw.rectangle([(0,y3),(900,y3_2)],fill='#EDFBFE') # 课程
            draw.rectangle([(0,y4),(900,y4_2)],fill='rgb(250,254,255)')# 照片
            draw.rectangle([(0,y5),(900,y5_2)],fill='rgb(250,254,255)') # 能力测评
            draw.rectangle([(0,y6),(900,y6_2)],fill='rgb(250,254,255)') # logo


            draw.rectangle([(pic0[0][0],pic0[0][1]),(pic0[0][2],pic0[0][3])],fill='#FFFFFF') #相框_课程
            draw.rectangle([(pic1[0][0],pic1[0][1]),(pic1[0][2],pic1[0][3])],fill='#FFFFFF') #相框_1
            draw.rectangle([(pic2[0][0],pic2[0][1]),(pic2[0][2],pic2[0][3])],fill='#FFFFFF') #相框_2
            draw.rectangle([(pic3[0][0],pic3[0][1]),(pic3[0][2],pic3[0][3])],fill='#FFFFFF') #相框_3
            draw.rectangle([(pic4[0][0],pic4[0][1]),(pic4[0][2],pic4[0][3])],fill='#FFFFFF') #相框_4
            
            return img                     
        def putImg(img):
            f_crs='I:\\每周实验课\\20200118-神奇的静电\\2020-01-18-神奇的静电-008.JPG'
            f_01='I:\\每周实验课\\20200118-神奇的静电\\2020-01-18-神奇的静电-008.JPG'
            f_02='I:\\每周实验课\\20200118-神奇的静电\\2020-01-18-神奇的静电-008.JPG'
            f_03='I:\\每周实验课\\20200118-神奇的静电\\2020-01-18-神奇的静电-008.JPG'
            f_04='I:\\每周实验课\\20200118-神奇的静电\\2020-01-18-神奇的静电-008.JPG'

            pic_crs=self.pic_resize(Image.open(f_crs),picWid)
            pic_01=self.pic_resize(Image.open(f_01),picWid)
            pic_02=self.pic_resize(Image.open(f_02),picWid)
            pic_03=self.pic_resize(Image.open(f_03),picWid)
            pic_04=self.pic_resize(Image.open(f_04),picWid)
            _pic_logo=Image.open('I:\\大智小超\\公共素材\\图片类\\大智小超新logo.png').convert('RGBA')
            pic_logo=_pic_logo.resize((350,int(350/2.76)))
            r,g,b,a=pic_logo.split()

            _pic_qrcode=Image.open('I:\\大智小超\\公共素材\\图片类\\大智小超二维码.jpg')
            pic_qrcode=_pic_qrcode.resize((130,130))


            print(pic_crs.size)
            img.paste(pic_crs,(pic0[1][0],pic0[1][1]))
            img.paste(pic_01,(pic1[1][0],pic1[1][1]))
            img.paste(pic_02,(pic2[1][0],pic2[1][1]))
            img.paste(pic_03,(pic3[1][0],pic3[1][1]))
            img.paste(pic_04,(pic4[1][0],pic4[1][1])) 
            img.paste(pic_logo,(50,1830),mask=a)
            img.paste(pic_qrcode,(700,1830))            
        def putTxt(img):            
            draw=ImageDraw.Draw(img)
            font_title = ImageFont.truetype(os.path.join(self.font_dir,'c:\\windows\\fonts\\腾祥金砖黑简.TTF'),70)
            font_1 = ImageFont.truetype(os.path.join(self.font_dir,self.default_font),60)
            font_2 = ImageFont.truetype(os.path.join(self.font_dir,self.default_font),50)
            font_3 = ImageFont.truetype(os.path.join(self.font_dir,self.default_font),40)
            font_4 = ImageFont.truetype(os.path.join(self.font_dir,self.default_font),30)
            font_sig = ImageFont.truetype(os.path.join(self.font_dir,'j:\\fonts\\2012DingYongKangYingBiKaiShuXinBan-2.ttf'),42)          
            font_5=os.path.join('j:\\fonts\\2012DingYongKangYingBiKaiShuXinBan-2.ttf')            
            draw.text((110,35), '乐高机器人课（初级）', fill = '#6AB34A',font=font_title)  #题目
            draw.text((300,160), '李XX', fill = '#6AB34A',font=font_1)  #姓名
            draw.text((530,160), '6岁', fill = '#6AB34A',font=font_1)  #年龄
            draw.text((120,290), crs_info[0], fill = '#6AB34A',font=font_4)  #课程           
#             draw.text((20,400), crs_info[1], fill = '#6AB34A',font=font_4)  #知识点
            self.put_txt_img(img,crs_info[1],410,[50,400],20,font_5,28)
            
            draw.text((100,650), '2020-07-28', fill = '#6AB34A',font=font_3)  #日期
            draw.text((530,650), '难度：'+crs_info[3], fill = '#6AB34A',font=font_3)  #难度
    #         draw.text((50,1490), '能力测评', fill = '#6AB34A',font=font_2)  #能力测评
    #         draw.text((50,1560), 'XX力', fill = '#6AB34A',font=font_3)  #XX力
    #         draw.text((50,1610), 'XX力', fill = '#6AB34A',font=font_3)  #XX力
    #         draw.text((50,1660), 'XX力', fill = '#6AB34A',font=font_3)  #XX力
    #         draw.text((50,1710), 'XX力', fill = '#6AB34A',font=font_3)  #XX力
    #         draw.text((290,1560), '李XX同学这段时间的各项能力都得到\n有效的提升。', fill = '#6AB34A',font=font_sig)  #评语
            script=expScript()
            self.put_txt_img(img,script,600,[60,1530],20,font_5,30,'add_2spaces')
    #     def put_txt_img(self,img,t,total_dis,xy,dis_line,font_size):

            draw.text((650,1800), '阿晓老师', fill = '#6AB34A',font=font_sig)  #签名
    
        para=basic_para()
        picWid,picX1,picX2,picX3,picX4,picY1,picY2,picY3,picY4,pic0,pic1,pic2,pic3,pic4= \
        para[0],para[1],para[2],para[3],para[4],para[5],para[6],para[7],para[8],para[9],para[10],para[11],para[12],para[13]
        
        
        img=basic_bg()    
        crs_info=read_excel()
        putImg(img)
        putTxt(img)        
        img.show()

    def pic_resize(self,pic,wid):
        w,h=pic.size
        r=h/w
        pic_resized=pic.resize((wid,int(wid*r)),Image.ANTIALIAS)
        return pic_resized    

    def split_txt(self,dis,font_size,txt_input):    
        txts=txt_input.split('\n')        
        logging.info(len(txts))
        spt=0
        for t in txts:        
            total_len=(self.char_len(t)+len(txts)*6)*font_size

            if total_len%dis==0:
                _spt=total_len/dis        
            else:
                _spt=total_len//dis+1
            spt=spt+_spt
            
        logging.info(spt)

        zi_per_line=int(dis//font_size)

        txt_L=[]
        logging.info(zi_per_line)        

        for txt in txts:
            t_n=[]
            for i in range(0,int(spt)):
                if i==0:
                    if txt[0:zi_per_line]:
                        t_n.append(txt[0:zi_per_line])
                else:
                    if txt[i*zi_per_line:(i+1)*zi_per_line]:
                        t_n.append(txt[i*zi_per_line:(i+1)*zi_per_line])
                        
                print(len(txt[i*zi_per_line:(i+1)*zi_per_line]))
        return txt_L
    
    def char_len(self,txt):
        len_s=len(txt)
        len_u=len(txt.encode('utf-8'))
        ziShu_z=(len_u-len_s)/2
        ziShu_e=len_s-ziShu_z
        total=ziShu_z+ziShu_e*0.5    
        return total

    def put_txt_img(self,img,t,total_dis,xy,dis_line,font,font_size,addSPC='None'):
        if addSPC=='add_2spaces':
            Line_Indent='yes'
        else:
            Line_Indent='no'

        txt=self.split_txt(total_dis,font_size,t)
        font_sig = ImageFont.truetype(font,font_size)
        num=len(txt)   
        draw=ImageDraw.Draw(img)
    
        n=0
        for t in txt:              
            m=0
            for tt in t:                   
                x,y=xy[0],xy[1]+(font_size+dis_line)*n
                if Line_Indent=='yes':   #首行缩进
                    if m==0:    
                        tt='  '+tt #首先前面加上两个空格
                        logging.info(str(x)+','+str(y))
                        logging.info(tt)
                        draw.text((x+font_size*1.5,y), tt, fill = '#6AB34A',font=font_sig) 
                    else:
                        logging.info(str(x)+','+str(y))
                        logging.info(tt)
                        draw.text((x,y), tt, fill = '#6AB34A',font=font_sig)  
                else: #首行不缩进
                    logging.info(str(x)+','+str(y))
                    logging.info(tt)
                    draw.text((x,y), tt, fill = '#6AB34A',font=font_sig)  
                m+=1
                n+=1

        
if __name__=='__main__':
    my=poster()
    my.PosterDraw('可以伸缩的夹子')        
```
